```python
def getphotodetial(filepath):
    answer = []
    fileList = os.listdir(filepath)
          # 获取path目录下所有文件
    for filename in fileList:
        print(filename)
        result = {}
        pathTmp = os.path.join(filepath,filename)  # 获取path与filename组合后的路径
        print(pathTmp)
        if os.path.isdir(pathTmp):
            getphotodetial(pathTmp)
        else:               
            minmane = pinyin(pathTmp,style=Style.NORMAL,strict=False)
            pname = ''.join([p[0] for p in minmane])
            #filename=''.join(re.findall(r'[a-zA-Z0-9]|\/|\.|:',pname))#只保留数字字母和两个符号/.
            result['行业调研题目']=filename.rstrip('.pdf')
            result['技术属性'] = 1
            result['艺术属性'] = 0
            result['文化属性'] = 0
            result['人本属性'] = 0
            result['商业属性'] = 1
            result['报告来源']=filepath.split('/')[-1]
            result['调研公司']=''
            result['报告时间']=''
            if "白皮书" in filename:
                result['报告类型']='白皮书'
            elif '前景'in filename:
                result['报告类型']='前景预测'
            elif '蓝皮书'in filename:
                result['报告类型']='蓝皮书'
            elif '调查报告'in filename:
                result['报告类型']='调查报告'
            elif '投资策略'in filename:
                result['报告类型']='投资策略'
            else:
                result['报告类型']='产业分析'

            result['摘要'] = None
            result['报告原文/原文链接'] = ''.join(re.findall(r'[a-zA-Z0-9]|\/|\.|:',pname))#只保留数字字母和两个符号/.
            result['报告类别'] = '行业深度调研'
            try:
                #f = pdfplumber.open(pathTmp)
                result['总页数']=len(pdfplumber.open(pathTmp).pages)
                text=[]
                if result['总页数']<15:
                    for i in range(result['总页数']):
                        page = pdfplumber.open(pathTmp).pages[i]   # 第一页的信息
                        text.append(page.extract_text()) 
                        result['textt']=text
                        result['zy']=str(text).split('。',1)[0].split('\\n')[-1]
                        result['zy2']=str(text).split('。',1)[0].replace('\\n','').replace("['",'')
                else :
                    for i in range(15):
                        page = pdfplumber.open(pathTmp).pages[i]   # 第一页的信息
                        text.append(page.extract_text()) 
                        result['textt']=text
                        result['zy']=str(text).split('。',1)[0].split('\\n')[-1]
                        result['zy2']=str(text).split('。',1)[0].replace('\\n','').replace("['",'')

            except:
                result['总页数']=None
                result['textt']=None
                result['zy']=None 
            print('get')
            answer.append(result)      # 将文件的大小添加到列表

                
        #if answer != []:
            df = pd.DataFrame(answer)
            lastname = str(filepath.split('/')[-1])
            name = 'ddd/'+lastname+'.xlsx'
            df.to_excel(name,index=False,columns=['行业调研题目','技术属性','艺术属性','文化属性','人本属性','商业属性','报告来源',
                                                 '报告类别','摘要','总页数','报告原文/原文链接','报告类别','zy','zy2','textt'])

```
